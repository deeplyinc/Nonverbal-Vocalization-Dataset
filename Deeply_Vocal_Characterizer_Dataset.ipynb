# Deeply Vocal Characterizer Dataset

```python
import librosa
import soundfile
import numpy as np
from IPython.display import Audio
```

#### Load & play example
the example loaded below is a coughing example.

```python
Fs = 16000

example_path = "Dataset/throat-clearing/1K0C_6_4_0_41_0_0_0.wav"
data, _ = librosa.core.load(example_path, sr = Fs)

soundfile.info(example_path) # returns the parameters of the audio file
```

```python
Audio(data, rate = Fs)
```

#### Convert the signal into mel spectrogram and visualize it

```python
import librosa.display
import matplotlib.pyplot as plt

spectrogram = librosa.feature.melspectrogram(y = data, sr = Fs, n_mels = 64, n_fft = 2048, hop_length = 512)

plt.figure(figsize = (10,8))

librosa.display.specshow(np.squeeze(spectrogram), x_axis = 'time', y_axis = 'mel', sr = Fs, cmap ='magma', fmax = 8000)

plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```
